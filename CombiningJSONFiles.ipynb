### Combines the JSON files into one data frame and create a SQL view to refereence the data frame

```
%python
path = "/mnt/bigdataanalytics/AccuWeatherHourly/*/*/*/*.json"
sdfWeatherData = spark.read.option("Multiline","false").json(path)
sdfWeatherData.createOrReplaceTempView("vwWeatherData")
```

### Activates the silverdbx database

```
%sql
USE silverdbx;
DROP TABLE IF EXISTS weatherdata;
```

### Removes the table from the data lake

```
%python
dbutils.fs.rm('/mnt/bigdataanalytics/silver/weatherdata',recurse=True)
```

### Creates the weatherdata delta table

```
%sql
CREATE TABLE weatherdata (
   EpochTime BIGINT
  ,HasPrecipitation BOOLEAN
  ,IsDayTime BOOLEAN
  ,Link VARCHAR(255)
  ,LocalObservationDateTime TIMESTAMP
  ,MobileLink VARCHAR(255)
  ,PrecipitationType VARCHAR(100)
  ,Fahrenheit DECIMAL(5,2)
  ,Celsius DECIMAL(5,2)
  ,WeatherIcon INT
  ,WeatherText VARCHAR(100)
)
USING DELTA 
LOCATION '/mnt/bigdataanalytics/silver/weatherdata'
```

### Populates the weatherdata delta table

```
%sql
INSERT INTO weatherdata (
   EpochTime
  ,HasPrecipitation
  ,IsDayTime
  ,Link
  ,LocalObservationDateTime
  ,MobileLink
  ,PrecipitationType
  ,Fahrenheit 
  ,Celsius
  ,WeatherIcon
  ,WeatherText
)

SELECT
   EpochTime
  ,HasPrecipitation
  ,IsDayTime
  ,Link
  ,LocalObservationDateTime
  ,MobileLink
  ,PrecipitationType
  ,Temperature['Imperial']['Value'] AS Fahrenheit 
  ,Temperature['Metric']['Value'] AS Celsius
  ,WeatherIcon
  ,WeatherText
FROM vwWeatherData
```

```
%sql
SELECT * FROM weatherdata ORDER BY EpochTime DESC LIMIT 10
```
